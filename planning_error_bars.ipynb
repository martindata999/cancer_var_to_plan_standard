```python
# Import Library
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import Data
data = pd.read_csv('data\diagnostic_actuals.csv')
                   
data.head()
```

```python
# Filter one planning metric, for testing

data_filtered = data[data['planning_ref'].isin(['E.B.26a', 'E.B.26d'])]

# Check filtered
data_filtered.head()
```

```python
# Filter numerator and denominator
numerator_data = data[data['measure_type'] == 'Numerator']
denominator_data = data[data['measure_type'] == 'Denominator']

# Merge numerator and denominator data on relevant columns
merged_data = pd.merge(numerator_data, denominator_data, on=['icb_code', 'org_code', 'dimension_name', 'source', 'planning_ref'], suffixes=('_num', '_den'))

# Calculate the percentage
merged_data['percentage'] = merged_data['metric_value_num'] / merged_data['metric_value_den']

# Select relevant columns for the new table
data_metrics = merged_data[['icb_code', 'org_code', 'dimension_name', 'source', 'planning_ref', 'percentage']]

# Display the result
data_metrics.head()
```

```python
# Get unique planning_refs
planning_refs = data_metrics['planning_ref'].unique()

for planning_ref in planning_refs:
    # Filter data for the current planning_ref
    data_filtered = data_metrics[data_metrics['planning_ref'] == planning_ref]
    
    # Group by org_code and calculate mean and standard error of the mean
    grouped = data_filtered.groupby('org_code').agg(
        mean_value=('percentage', 'mean'),
        count=('percentage', 'count'),
        std_dev=('percentage', 'std')
    ).reset_index()
    
    # Calculate standard error
    grouped['std_error'] = grouped['std_dev'] / np.sqrt(grouped['count'])
    
    # Extract data for plotting
    org_codes = grouped['org_code']
    values = grouped['mean_value']
    errors = grouped['std_error']
    
    # Create the error bar chart
    plt.figure(figsize=(10, 6))
    plt.errorbar(values, org_codes, xerr=errors, fmt='o', ecolor='r', capsize=5, linestyle='None', marker='s', markersize=8)
    
    # Add markers for each org_code for end of year plan goal
    unique_org_codes = data_filtered['org_code'].unique()
    for org_code in unique_org_codes:
        plt.scatter(0.2, org_code, label=org_code, marker='x')
    
    # Add labels and title
    plt.xlabel('Mean Performance')
    plt.ylabel('Org Code')
    plt.title(f'Error Bar Chart by Org Code with Standard Error for {planning_ref}')
    plt.grid(True)
    
    # Show the plot
    plt.show()
```
